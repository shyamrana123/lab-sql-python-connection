```python
#Write a Python function called rentals_month that retrieves rental data for a given month and year (passed as parameters) from the Sakila database as a Pandas DataFrame. 
# The function should take in three parameters:
    #engine: an object representing the database connection engine to be used to establish a connection to the Sakila database.
    #month: an integer representing the month for which rental data is to be retrieved.
    #year: an integer representing the year for which rental data is to be retrieved.
#The function should execute a SQL query to retrieve the rental data for the specified month and year from the rental table in the Sakila database, 
# and return it as a pandas DataFrame.
```

```python
def rentals_month(engine, month, year):
    
    query = (f"""SELECT * FROM rental
             WHERE MONTH(rental_date)= {month} AND YEAR(rental_date) = {year}""")
    engine.execute(query)
    results = engine.fetchall()

    column_names = [element[0] for element in engine.description]
    return(pd.DataFrame(results, columns = column_names))
```

```python
rentals_month(engine, month, year)
```

```python
# function to get the two months
def getdate():
    months_years = []
    for i in range(1,3):
        month = int(input(f"Please enter the Month{i} you want to get the data of: "))
        year = int(input(f"Please enter the Year{i} you want to get the data of: "))
        month_year = [month,year]
        months_years.append(month_year)
    return months_years
```

```python
#Calling the function to get the months from user
user_months_years = getdate()
```

```python
# Loop to create and store DataFrames
dataframes = {}
for i in range(len(user_months_years)):
   
    month = user_months_years[i][0]
    year = user_months_years[i][1]
    
    results = rentals_month(engine, month, year)

    # Call the function and store the result in the dictionary
    dataframes[f'df{i+1}'] = rental_count_month(results)
    print(f'df{i+1}')
    display(dataframes[f'df{i+1}'])

df1 = dataframes['df1']
df2 = dataframes['df2']
```

```python
def compare_rentals(df1, df2):
    # Merging the two DataFrames on 'customer_id'
    merged_df = pd.merge(df1, df2, on='customer_id')
    
    # Adding a new column with the difference in rentals between the two DataFrames
    merged_df['difference'] = merged_df[f'rentals_{user_months_years[0][0]}_{user_months_years[0][1]}']- merged_df[f'rentals_{user_months_years[1][0]}_{user_months_years[1][1]}']
        
    # Optionally return the DataFrame for further use
    return merged_df
```
